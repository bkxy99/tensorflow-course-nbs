```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
tf.random.set_seed(42)
```

```python
# Features
X = np.array([-7.0, -4.0, -1.0, 2.0, 5.0, 8.0, 11.0, 14.0])

y = np.array([3.0, 6.0, 9.0, 12.0, 15.0, 18.0, 21.0, 24.0])

plt.scatter(X, y)
```

```python
# Create the model with Sequential API
model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(1)
])

model_1.compile(
    optimizer=tf.keras.optimizers.SGD(),
    loss="mae",
    metrics=["mae"]
)

model_1.fit(tf.expand_dims(X, axis=1), y, epochs=100)
```

```python
model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation="relu"),
    tf.keras.layers.Dense(1)
])

model_2.compile(
    optimizer=tf.keras.optimizers.SGD(),
    loss="mae",
    metrics=["mae"]
)

model_2.fit(tf.expand_dims(X, axis=1), y, epochs=100)
```

```python
X = tf.constant(np.arange(1, 100))
y = 2 * X + 10
```

```python
plt.plot(X, y)
```

```python
X_train = X[:80]
y_train = y[:80]

X_test = X[80:]
y_test = y[80:]
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(X_train, y_train, c="b", label="Training set")
plt.scatter(X_test, y_test, c="g", label="Testing set")
plt.legend()
```
